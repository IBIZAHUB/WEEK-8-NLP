## Practical Task 1: Sentiment Analysis with Hugging Face Transformers

First, we need to install the `transformers` library, which provides access to pre-trained models for various NLP tasks, including sentiment analysis.

```python
pip install transformers
```

Now, let's import the necessary components from the `transformers` library: `pipeline` for easily performing tasks, and `AutoTokenizer`/`AutoModelForSequenceClassification` for more control over the model loading process.

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

# We'll use a pre-trained model specifically designed for sentiment analysis.
# 'distilbert-base-uncased-finetuned-sst-2-english' is a good choice for general sentiment.
model_name = "distilbert-base-uncased-finetuned-sst-2-english"

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Create a sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

Next, we'll define our positive, negative, and neutral reviews and then use the `sentiment_analyzer` pipeline to predict their sentiment.

```python
# Define the reviews
positive_review = "This movie was absolutely fantastic! I loved every minute of it."
negative_review = "The service was terrible and the food was cold. A truly disappointing experience."
neutral_statement = "The capital of France is Paris."

# Perform sentiment analysis
results_positive = sentiment_analyzer(positive_review)
results_negative = sentiment_analyzer(negative_review)
results_neutral = sentiment_analyzer(neutral_statement)

# Print the results
print(f"Positive Review: '{positive_review}'")
print(f"Sentiment: {results_positive}\n")

print(f"Negative Review: '{negative_review}'")
print(f"Sentiment: {results_negative}\n")

print(f"Neutral Statement: '{neutral_statement}'")
print(f"Sentiment: {results_neutral}\n")
```

### Interpretation of Results

The output from the sentiment analysis pipeline provides a `label` (either 'POSITIVE' or 'NEGATIVE' for this specific model) and a `score` representing the confidence of the model in that label.

*   **Positive Review:** The model correctly identifies the positive review as `POSITIVE` with a high confidence score.
*   **Negative Review:** The model correctly identifies the negative review as `NEGATIVE` with a high confidence score.
*   **Neutral Statement:** The model's classification of the neutral statement as either `POSITIVE` or `NEGATIVE` is generally based on the likelihood it assigns to each, even if the text doesn't express strong emotion. In this case, `distilbert-base-uncased-finetuned-sst-2-english` is trained on a dataset where 'neutral' examples might be implicitly mapped to one of the two polarities if no strong emotion is present. The score for neutral statements is usually less extreme (closer to 0.5 for a binary classification) than for strongly emotional texts.

## Practical Task 2: Text Generation with GPT-2

For this task, we will use the `transformers` library again to leverage the pre-trained GPT-2 model. The library should already be installed from the previous task.

```python
from transformers import pipeline

# Initialize the text generation pipeline with GPT-2
generator = pipeline('text-generation', model='gpt2')
```

Now, let's create a cell where you can input a prompt, and the GPT-2 model will generate a text continuation. You can modify the `prompt` variable and re-run the cell to experiment with different inputs.

```python
# Accept user input for a prompt
prompt = "Once upon a time, in a land far, far away,"

# Generate text continuations
# You can adjust max_length and num_return_sequences for different output lengths and numbers of generations
generated_text = generator(prompt, max_length=100, num_return_sequences=1, do_sample=True, temperature=0.9)

# Display the generated text
print("--- Your Prompt ---")
print(prompt)
print("\n--- Generated Text ---")
print(generated_text[0]['generated_text'])
```

Try changing the `prompt` in the cell above to different sentences. For example:
